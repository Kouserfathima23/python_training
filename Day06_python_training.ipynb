```python
students = ["kouser","fathima","khatija","mariam","zainab","kulsum","nisar"]
print(students)
```

```python
#appreciating khatija for good work!
message = f"keep up the good work,{students[2].title()}"
print(message)
```

```python
#appreciating nisar for good work!
message = f"keep up the good work,{students[6].title()}"
print(message)
```

```python
#to send message of appreciation to all at once!
for x in students:
    print(f"keep the good work, {x.title()}")
    
```

```python
#to send message of appreciation to all at once!
for x in students:
    print(f"keep the good work, {x.title()}")
    print(f"will be looking forward on meeting you soon for the new launch of our product, {x.title()}")
```

```python
#to make the message look clean ,we will use /n (space in the end )
for x in students:
    print(f"keep the good work, {x.title()}")
    print(f"will be looking forward on meeting you soon for the new launch of our product, {x.title()}.\n")
```

```python
#NOTE:indentation is important in loop statments if not it won't be excecuted and print statements within the loop will be excecuted to all ,outside the print statemnts will be excecuted only once in the end.
```
