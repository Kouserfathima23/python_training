```python
#continuation with list datatypes
students = ['kouser','fathima','khatija','mariam','faiyaz','sharukh','doja cat','hyder']
print(students)
```

```python
#to sort them in order
print(sorted(students))
```

```python
#permenant sort
students.sort()
print(students)
```

```python
#to get opposite orders of given list
students.reverse()
print(students)
```
